Impact of COVID-19 on Tourism (2020)

```python
commodities = ['Traveler accommodations', 'Food and beverage services', 'Transportation', 'Recreation, entertainment, and shopping']
selected_data = []
for commodity in commodities:
    commodity_data = tourism[tourism['Commodity'] == commodity]
    growth_2020 = tourism['contribution2020'].values[0]  
    selected_data.append((commodity, growth_2020))
selected_df = pd.DataFrame(selected_data, columns=['Commodity', '2020 Growth Rate'])
selected_df
```

```python
tourism_economy = pd.read_csv("../Data/tourism1clean.csv") 

growth_cols = [col for col in tourism_economy.columns if col.startswith("Annual growth")]
tourism_economy_growth = tourism_economy.melt(id_vars=["Commodity"], value_vars=growth_cols,
                    var_name="Year", value_name="Annual Growth")
tourism_economy_growth["Year"] = tourism_economy_growth["Year"].str.extract(r'(\d{4})')

contrib_cols = [col for col in tourism_economy.columns if col.startswith("contribution")]
tourism_economy_contrib = tourism_economy.melt(id_vars=["Commodity"], value_vars=contrib_cols,
                     var_name="Year", value_name="Contribution")
tourism_economy_contrib["Year"] = tourism_economy_contrib["Year"].str.extract(r'(\d{4})')

output_cols = [col for col in tourism_economy.columns if col.startswith("RealOutput")]
tourism_economy_output = tourism_economy.melt(id_vars=["Commodity"], value_vars=output_cols,
                    var_name="Year", value_name="Real Output (in millions)")
tourism_economy_output["Year"] = tourism_economy_output["Year"].str.extract(r'(\d{4})')

tourism_economy_long = tourism_economy_growth.merge(tourism_economy_contrib, on=["Commodity", "Year"])
tourism_economy_long = tourism_economy_long.merge(tourism_economy_output, on=["Commodity", "Year"])

for col in ["Annual Growth", "Contribution", "Real Output (in millions)"]:
    tourism_economy_long[col] = tourism_economy_long[col].astype(str).str.replace("−", "-", regex=False)
    tourism_economy_long[col] = pd.to_numeric(tourism_economy_long[col], errors="coerce")

tourism_economy_long["Year"] = tourism_economy_long["Year"].astype(int)
tourism_economy_long.head()
```

```python
tourism_economy_long.to_csv('tourismeconomy.csv')
```
